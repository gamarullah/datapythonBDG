```python
# LIBRARY
import matplotlib.pyplot as plt
import numpy as np
```

```python
def function(x,y):
    return (x+y)*2
function(10,5)
```

```python
def square(x):
    """square a number"""
    return x ** 2

for N in range(0, 5):
    print(N, "squared is", square(N))
```

```python
# Fixing random state for reproducibility
np.random.seed(18112018)


N = 100
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```
